# R / RStudio

## Packages & Environment

Install package

    install.packages("package.name")
    
Load pacakage

    library(package)
    
Get Working Directory

    getwd
    
Set Working Directory

    setwd("path")
    
## Lists

Initialize an empty list. List can contain elements of different types

    L<-list()
    
Get number of row, columns, names of columns, head

    nrow(L)
    ncol(L)
    names(L)
    head(L)
    
Get a summary of the data (Min, 1st Q, Median, Mean, 3rd Q, Max)

    summary(L)
    
Bind a columnwise or rowwise

    cbind(L1,L2)
    rbinbd(L1,L2)
    
## Vector

Initialize an empty vector. Vector can only contain one type of element

    V<-c()
    
## Matrix

Initialize an empty matrix

    M <- matrix(nrow = n,ncol = m)
    
## DataFrame

Declare a DataFrame

    df<- data.frame(var1= c(data1,data2 ...), var2=c(data1,data2 ...), ...)
    OR
    df <- data.frame (rownames = c(namerow1, ...), colnames = c(namecol1, ...), c(data1, ...), c(data1, ...) ...)

Get structure or summary of DataFrame

    str(df)
    summary(df)
    
Get subset of dataframe

    df$colname1     # $ column with name colname1
    df[,1]          #first column
    df[,-c(4,7,10)] #df without column 4 ; 7 and 10
    df[1:10,]        #first ten lines of df
