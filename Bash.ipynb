## Navigating

Print Working Directory

    pwd
    
List element (sorted) in directory
    
    ls (| sort) 
    
Enter a folder

    cd directoryName
    cd ..           #goes to parent directory
    cd              #goes to home directory
    
Path tricks

    directory/*          #represents all the files in a directory
    .                    #represents the working directory
    ~                    #represents your home directory
    pushd directoryName  #saves the directory to a the stack
    cd ~1                #goes to the lastly stacked directory
    dirs -v              #shows the directory stack
    
    
Make a directory

    mkdir directoryName
    
Remove a directory (recursively)

    rmdir -r directoryName
    
Remove a file (or a directory recursively)

    rm fileName
    rm -r directoryName
    
copy a file (or a directory recursively)

    cp fileSource fileDestination
    cp -r directorySource directoryDestination
    
Move a file or a folder

    mv fileSource fileDestination

## Authorization

The authorization of files are listed when you call ls -la  <br>
It shows a code of 10 characters. <br>
First is for the type of file : d = directory, - = file, l = symbolic link
Next triplet are for read (r) write (w) and exectute (x) authorization of all the users

    type owner group other
      -   rwx   rwx   rwx    File.exa
    #This file can be listed, modified and executed by any user
    
    drwx--x--x MyDir
    #Every user can to pass through MyDir (to access a subdir for example) but only the owner can list and modify it
    
Change authorization for File.exa <br>
First argument is for the users affected : u = owner, g = group, o = other user, a = all <br>
Second argument is + to give authorization or - to withdraw authorization <br>
Last argument is the type of authorization
         
    chmod g+rwx File.exa
    #the owner gives read write and execute authorization to all his group

# SSH

Generate a publi/private key

    ssh-keygen

Connect to an SSH server

    ssh username@ssh.server
    
Exit the server
    
    exit
    
Copy from and on a server

    scp "file_to_copy_path" username@ssh.server:"paste_folder"  #Send a file to an ssh adress
    scp username@ssh.server:"file_to_copy_path" "paste_folder"  #Copy a file from an ssh adress
    
    scp -r "folder_to_copy" username@ssh.server:"paste_folder"  #Send a folder to an ssh adress
    scp -r username@ssh.server:"folder_to_copy" "paste_folder"  #Copy a folder from an ssh adress
    
    

# Accelerating tips

Save historic from all terminal

    shopt -s histappend 
    
Repeat last instruction as superuser
    
    sudo !!
    
Reuse arguments from previous command

    !*
    
Get back to last directory

    cd -
    
Shortcuts

    ctrl + _ (undo)
    ctrl + t (swap two characters)
    ALT + t (swap two words)
    ALT + . (prints last argument from previous command)
    ctrl + arrow (move forward a word)
    ALT + f (move forward a word)
    ALT + b (move backward a word)
    ctrl + e (move cursor to end)
    ctrl + a (move cursor to start)
    ctrl + xx (move to the opposite end of the line)
    ctrl + u (cuts everything before the cursor)
    ctrl + k (cuts everything after the cursor)
    ctrl + y (pastes from the buffer)
    ctrl + l (clears screen)s
