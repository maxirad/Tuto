### Job submission

Set a job-name
Set your account
in a .sh file

    #!/bin/bash
    #SBATCH --job-name=180911_Lim
    #SBATCH --account=def-banire
    #SBATCH --time=45:30:00
    #SBATCH --mem=515000M
    #SBATCH --output=/project/def-banire/midou/bioinformatics/myout.out
    #SBATCH --error=/project/def-banire/Labobioinfo/Jobs/180911_Lim/results/siH_hisat2.err
    #SBATCH --mail-user=youremail
    #SBATCH --mail-type=ALL

    bash instruction
    module load python/3-6-3
    python myscript.py      #implicitly in the same dir or give path

Launch a job

    sbatch jobfile.sh

See if you're job is done
    
    squeue -u userid
    squeue -u userid -t "PENDING"
    squeue -u userid -t "RUNNING"
    
Cancel a job
    
    scancel job_id
    scancel -u userid    #(all job from a user)

    module spider python

    module load python/2.7.14

    pip install --user numpy
