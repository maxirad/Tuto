# Basic

## Install Git & check version

    sudo apt-get install git
    git --version

## Configuration

    git config --global user.name <YourName>
    git config --global user.email <YourEmail>
    
## Manage files locally
#### Create a new git folder

    git init <FolderName>    
#### Create a new file and save it in the staging area
    
    touch <file.ext>
    git add <file.ext>    
#### Remove the file from the staging area
    
    git reset <file.ext>    
#### Commit the staging area
    
    git commit -m "commit information"    
#### Get status of commitment

    git status
    git log   
#### Get difference between :
working directory and staging area

    git diff
staging area and repository
    
    git diff --staged
Your local branch and the remote branch

    git diff <local branch> <remote>/<remote branch>

## Manage Files Remotely
#### Add remote
Check for remote

    git remote --v
Add a remote (named origin at URL)

    git remote add <remote> URL    
#### Push Code
Push a master to origin (-u sets it as a reference for a next git pull)

    git push <remote> master
Push a branch to origin (-u sets it as a reference for a next git pull)

    git push <remote> <branch>   
Push all branches to origin

    git push --all    
#### Fetch Code
Get code from the origin, pull = fetch + merge

    git fetch
    git pull   
#### Show where the master is

    git show master

## Undo things
Write a commit that undoes commit-id (if you want to undo the last you can use HEAD)

    git revert <commit-id>
    (git push <remote> <branch>)    
Get back to a previous version of the repository (after commit-id)

    git reset --hard <commit-id>
    (git push -f <remote> <branch>) #You will have to force push as you go back in time

## Branching
Show branches (-a to show the branches of your remote)

    git <branch>
    tree .git/refs    
Create a new branch (before or N commits before master)

    git branch <newBranch>
    git branch <newPrevious> master^
    git branch <newNprevious> master~N    
Checkout a new branch (-b lets you create a new branch whilst checkout on it)
    
    git checkout <newBranch>
Delete a branch locally and remotely

    git branch -d <branch>
    git push --delete <remote> <branch>

## Ignore some files

Create the file that manage the ignored files

    touch .gitignore
    gedit .gitignore
Typical gitignore file <br>
https://gist.github.com/octocat/9257657
    
Clean your repo cache (if you added stuff you then want to ignore) <br>
http://www.codeblocq.com/2016/01/Untrack-files-already-added-to-git-repository-based-on-gitignore/

    git rm -r --cached .
    
## Useful aliases

Add these aliases in your ~/.bashrc

    # Git
    alias g='git'
    alias gs='git status'
    alias ga='git add'
    alias gc='git commit --m'
    alias gp='git push origin'
    alias gb='git branch'
    alias gco='git checkout'
    alias gcom='git checkout master'
    alias gl='git log'
    alias gpl='git pull'
Then the only thing you need to push is to do

    ga [file1] [file2] ...
    gc "[your message]"
    gp
